## Calculate Average Temp and Deviation

```python
weather_data = pd.read_csv(".\data\weather_jan.csv", encoding='cp1250', index_col=0, parse_dates=True, dayfirst=True)

average = weather_data.groupby([ weather_data.index.day, weather_data.index.time])['T [C]'].mean()

weather_data = weather_data.join(pd.DataFrame(average).rename(columns={"T [C]":"AvrT"}), 
                                           on=[weather_data.index.day, weather_data.index.time], how = 'left')

std = weather_data.groupby([ weather_data.index.day, weather_data.index.time])['T [C]'].std()
print "Mean standard deviation: ", std.mean()

weather_data = weather_data.join(pd.DataFrame(std).rename(columns={"T [C]":"StdT"}), 
                                           on=[weather_data.index.day, weather_data.index.time], how = 'left')

weather_data
_ = weather_data['2014-01-01 00:00:00':'2014-01-01 23:30:00'][['AvrT','StdT']].plot(kind='bar', figsize=(18,4), 
    title="Average Temperature Over One Day (1.1.2014)", yerr='StdT', ylim=(-10,10))
```

From the historical data (from 1994 to 2014), we calculate average Temperature for January. Bellow we can see average temperature for one day (1. Jan 2014), with standard deviatons. Mean standard deviation for January is 4.2 C.

## Comapring Average Temperatures with Standard Deviatons.

```python
weather_data = pd.read_csv(".\data\weather_jan.csv", encoding='cp1250', index_col=0, parse_dates=True, dayfirst=True)

mean = weather_data.groupby([ weather_data.index.day, weather_data.index.time])['T [C]'].mean()
std = weather_data.groupby([ weather_data.index.day, weather_data.index.time])['T [C]'].std()

weather_data = weather_data.join(pd.DataFrame(mean).rename(columns={"T [C]":"AvrT"}), 
                                           on=[weather_data.index.day, weather_data.index.time], how = 'left')

weather_data = weather_data.join(pd.DataFrame(std).rename(columns={"T [C]":"StdT"}), 
                                           on=[weather_data.index.day, weather_data.index.time], how = 'left')

plt.figure(figsize=(20,6))

for i, group in weather_data.groupby(weather_data.index.year):
    ax = group['T [C]'].plot(use_index=False, legend=True, label=i)

weather_data[0:group.shape[0]]['AvrT'].plot(use_index=False, linewidth=3.0)
plt.fill_between(np.arange(len(group)),
             (weather_data['AvrT'] + weather_data['StdT']/2)[0:group.shape[0]].values,
             (weather_data['AvrT'] - weather_data['StdT']/2)[0:group.shape[0]].values, 
             alpha=0.5, color='b', label="Std")
#plt.plot(weather_data[0:group.shape[0]].index,
#         weather_data['AvrT'][0:group.shape[0]], linewidth=3.0)

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

Same as previous, only with standard deviation.

## Temperature Heatmap

```python
import seaborn as sns 

weather_data = pd.read_csv(".\data\weather_jan.csv", encoding='cp1250', parse_dates=[0], dayfirst=True)

weather_data["DayTime"] = weather_data['datetime'].apply(lambda x: x.strftime('%b-%dT%H:%M:%S'))
weather_data["Year"] = weather_data['datetime'].apply(lambda x: x.year)
#weather_data.pivot(weather_data['DayTime'], weather_data['Year'], 'T')

temperature = pd.DataFrame.pivot_table(weather_data, index=['DayTime'], columns=['Year'], values='T [C]')
#temperature["Average"] = temperature.apply(np.mean, axis=1)
diff = temperature.sub(temperature.apply(np.mean, axis=1), axis=0)


# Format tick
xticks = diff.index
keptticks = xticks[::int(len(xticks)/10)]
xticks = ['' for y in xticks]
xticks[::int(len(xticks)/10)] = keptticks

# Display heatmap
plt.figure(figsize=(10,7))
sns.heatmap(temperature.T, linewidths=0, cmap="RdBu_r" ,xticklabels=xticks) # Plot temperature heatmap
#sns.heatmap(diff.T, linewidths=0, cmap="RdBu_r" ,xticklabels=xticks) # Plot temperature difference heatmap
```

Visualizing January temperatures over years from 1994 to 2014. It is clear, how some year like 2014 were hotter than usually, whicle some years like 2003 were colder.

## Boxplot

```python
weather_data = pd.read_csv(".\data\weather_jan.csv", encoding='cp1250', parse_dates=[0], dayfirst=True)

weather_data["DayTime"] = weather_data['datetime'].apply(lambda x: pd.to_datetime(x.strftime('%b-%dT%H:%M:%S'), format="%b-%dT%H:%M:%S"))
weather_data["Year"] = weather_data['datetime'].apply(lambda x: x.year)
#weather_data.pivot(weather_data['DayTime'], weather_data['Year'], 'T')

temperature = pd.DataFrame.pivot_table(weather_data, index=['DayTime'], columns=['Year'], values='T [C]')
temperature["Average"] = temperature.apply(np.mean, axis=1)

sns.set_style("whitegrid")
plt.figure(figsize=(15,6))
plt.xticks(rotation=45)
#plt.axhline(mean.mean(), linewidth=1.0, alpha=0.5, label="Mean", zorder=-1)
#plt.axhspan(mean.mean() - std.mean(), mean.mean() + std.mean(), alpha=0.2, label="Std", zorder=-1)
#sns.boxplot(temperature.loc[:, range(1994,2015)])
sns.boxplot(temperature.loc[:, temperature.columns])
plt.ylabel("January Temperature")
plt.show
#plt.legend()
sns.despine(left=True)
```

## Violinplot

```python
weather_data = pd.read_csv(".\data\weather_jan.csv", encoding='cp1250', parse_dates=[0], dayfirst=True)

weather_data["DayTime"] = weather_data['datetime'].apply(lambda x: pd.to_datetime(x.strftime('%b-%dT%H:%M:%S'), format="%b-%dT%H:%M:%S"))
weather_data["Year"] = weather_data['datetime'].apply(lambda x: x.year)
#weather_data.pivot(weather_data['DayTime'], weather_data['Year'], 'T')

temperature = pd.DataFrame.pivot_table(weather_data, index=['DayTime'], columns=['Year'], values='T [C]')
temperature["Average"] = temperature.apply(np.mean, axis=1)
plot_vals = [v.dropna() for k, v in temperature.iteritems()] 

sns.set_style("whitegrid")
plt.figure(figsize=(15,7))
plt.xticks(rotation=45)
plt.axhline(mean.mean(), linewidth=1.0, alpha=0.5, label="Mean", zorder=-1)
plt.axhspan(mean.mean() - std.mean(), mean.mean() + std.mean(), alpha=0.2, label="Std", zorder=-1)
sns.violinplot(plot_vals, names=temperature.columns)
plt.ylabel("January Temperature")
plt.legend()
plt.show
sns.despine(left=True)
```

Even more descriptive than previous is so called "violin" (or sometimes "viola") plot. Similar as box plot, the violin plot by default plots the median, along with the 25th and 75th percentile. But the shape of object tells us the distribution od data. In the background, mean temperature and its standard deviation is displayed for comparison. 

Again, we can observe how temperatures vary from year to year. For example, with displaying heatmap we have allready showed how 2014 was hotter year than average, and 2003 was colder. This can be seen in this plot as well. Even more, we can also see the distribution of temperature data from this plot.
